```python
"""
GENERADOR DE DATASET PARA NN2: OPTIMIZACIÓN DE CONEXIONES RBS
=============================================================
Este código genera automáticamente un dataset de 300+ combinaciones de parámetros RBS
para entrenar la red neuronal NN2 que predice el desempeño sísmico.

Autor: [Tu nombre]
Tesis: Optimización Automatizada de Conexiones Sismorresistentes mediante Redes Neuronales
"""

import openseespy.opensees as ops
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import product
import time
import warnings
warnings.filterwarnings('ignore')

class RBSDatasetGenerator:
    def __init__(self):
        # Propiedades geométricas base de la viga (según tu código original)
        self.L_beam = 2.0          # Longitud total de la viga (m)
        self.h = 0.5               # Altura total de la sección (m)
        self.tw = 0.01             # Espesor del alma (m)
        self.tf = 0.02             # Espesor del ala (m)
        self.bf_total = 0.2        # Ancho total del ala (m)
        
        # Propiedades del acero A36
        self.Fy = 250e6            # Límite elástico (Pa)
        self.Fu = 400e6            # Tensión de rotura (Pa)
        self.E = 200e9             # Módulo de elasticidad (Pa)
        
        # Factores para cálculo de Mp
        self.Ry = 1.3              # Factor de resistencia
        self.Cpr = min((self.Fy + self.Fu)/(2*self.Fy), 1.2)
        
        # Lista para almacenar resultados
        self.dataset = []
        
    def calculate_plastic_moment(self, bf_actual):
        """Calcula el momento plástico para un ancho de ala dado"""
        Zx = (0.25 * self.tw * (self.h**2)) + (bf_actual * self.tf * (self.h + self.tf))
        Mp = self.Fy * Zx * self.Ry
        return Mp
    
    def define_parameter_ranges(self):
        """Define los rangos de parámetros RBS según AISC358 y variaciones adicionales"""
        # Rangos base según AISC358
        a_base = np.arange(0.5, 0.8, 0.05) * self.h  # 0.5h a 0.75h
        b_base = np.arange(0.1, 0.3, 0.02) * self.bf_total  # 0.1bf a 0.25bf  
        c_base = np.arange(0.65, 0.85, 0.03) * self.h  # 0.65h a 0.8h
        
        # Rangos extendidos para exploración adicional
        a_extended = np.arange(0.8, 1.2, 0.1) * self.h
        b_extended = np.arange(0.3, 0.45, 0.05) * self.bf_total
        c_extended = np.arange(0.85, 1.0, 0.05) * self.h
        
        # Combinar rangos
        a_values = np.concatenate([a_base, a_extended])
        b_values = np.concatenate([b_base, b_extended])  
        c_values = np.concatenate([c_base, c_extended])
        
        return a_values, b_values, c_values
    
    def validate_rbs_geometry(self, a, b, c):
        """Valida que la geometría RBS sea físicamente factible"""
        # Verificaciones básicas
        if b >= self.bf_total/2:  # No puede reducir más del 50% del ala
            return False
        if a + 2*c >= self.L_beam * 0.8:  # RBS no puede ocupar más del 80% de la viga
            return False
        if c < 0.1:  # Longitud mínima de transición
            return False
        if a < 0.1:  # Distancia mínima desde el apoyo
            return False
            
        return True
    
    def simulate_rbs_performance(self, a, b, c, max_steps=1000):
        """
        Simula el comportamiento de una viga RBS y extrae métricas de desempeño
        Versión optimizada del código original
        """
        try:
            # Limpiar modelo anterior
            ops.wipe()
            
            # Geometría RBS
            bf_rbs = self.bf_total - 2 * b
            
            # Crear modelo
            ops.model("basic", "-ndm", 2, "-ndf", 3)
            ops.node(1, 0, 0)
            ops.node(2, a, 0) 
            ops.node(3, a + 2*c, 0)
            ops.node(4, self.L_beam, 0)
            ops.fix(1, 1, 1, 1)
            
            # Material Steel02
            ops.uniaxialMaterial("Steel02", 1, self.Fy, self.E, 0.03, 15, 0.925, 0.15)
            
            # Sección con fibras (RBS)
            ops.section("Fiber", 1)
            d_rbs = self.h - 2 * self.tf
            ops.patch("rect", 1, 2, 8, -self.tw/2, -d_rbs/2, self.tw/2, d_rbs/2)
            ops.patch("rect", 1, 8, 2, -bf_rbs/2, d_rbs/2, bf_rbs/2, d_rbs/2 + self.tf)
            ops.patch("rect", 1, 8, 2, -bf_rbs/2, -d_rbs/2 - self.tf, bf_rbs/2, -d_rbs/2)
            
            # Transformación e integración
            ops.geomTransf("PDelta", 1)
            ops.beamIntegration("Lobatto", 1, 1, 5)
            
            # Elementos
            ops.element("forceBeamColumn", 1, 1, 2, 1, 1)
            ops.element("forceBeamColumn", 2, 2, 3, 1, 1) 
            ops.element("forceBeamColumn", 3, 3, 4, 1, 1)
            
            # Análisis inicial
            ops.timeSeries("Linear", 1)
            ops.pattern("Plain", 1, 1)
            ops.load(4, 0, 1e-6, 0)
            
            ops.system("BandGeneral")
            ops.constraints("Plain")
            ops.numberer("RCM")
            ops.test("NormDispIncr", 1e-3, 50)
            ops.algorithm("Newton")
            ops.integrator("LoadControl", 1.0)
            ops.analysis("Static")
            ops.analyze(1)
            
            # Protocolo de carga cíclica simplificado (para generar dataset más rápido)
            drift_levels = [0.00375, 0.0075, 0.015, 0.03, 0.05]  # Niveles reducidos
            cycles_per_level = [2, 2, 1, 1, 1]  # Menos ciclos por nivel
            steps_per_half_cycle = 10  # Pasos reducidos
            
            # Listas para resultados
            disp_list = [ops.nodeDisp(4, 2)]
            moment_list = [ops.eleForce(2, 2)]
            step_count = 1
            
            ops.loadConst("-time", 0.0)
            
            # Bucle de carga cíclica
            for i, drift in enumerate(drift_levels):
                if step_count > max_steps:  # Límite de pasos para evitar simulaciones muy largas
                    break
                    
                num_cycles = cycles_per_level[i]
                target_disp_amplitude = drift * self.L_beam
                
                for cycle in range(num_cycles):
                    for sign in [-1, 1]:
                        target_disp = target_disp_amplitude * sign
                        current_disp = ops.nodeDisp(4, 2)
                        disp_to_travel = target_disp - current_disp
                        increment = disp_to_travel / steps_per_half_cycle
                        
                        ops.integrator("DisplacementControl", 4, 2, increment)
                        ops.analysis("Static")
                        
                        for step in range(steps_per_half_cycle):
                            ok = ops.analyze(1)
                            if ok != 0:
                                # Intentar algoritmo modificado
                                ops.algorithm("ModifiedNewton", "-initial")
                                ok = ops.analyze(1)
                                if ok != 0:
                                    break
                                ops.algorithm("Newton")
                            
                            disp_list.append(ops.nodeDisp(4, 2))
                            moment_list.append(-ops.eleForce(2, 2))
                            step_count += 1
                            
                            if step_count > max_steps:
                                break
                        
                        if ok != 0 or step_count > max_steps:
                            break
                    if ok != 0 or step_count > max_steps:
                        break
                if ok != 0 or step_count > max_steps:
                    break
            
            # Calcular métricas de desempeño
            if len(disp_list) > 1 and len(moment_list) > 1:
                max_rotation = max(abs(d) for d in disp_list) / self.L_beam
                max_moment = max(abs(m) for m in moment_list)
                
                # Métricas adicionales
                energy_dissipated = self._calculate_hysteretic_energy(disp_list, moment_list)
                ductility = self._calculate_ductility(disp_list, moment_list)
                
                return {
                    'max_rotation': max_rotation,
                    'max_moment': max_moment,
                    'energy_dissipated': energy_dissipated,
                    'ductility': ductility,
                    'convergence': True,
                    'steps_completed': step_count
                }
            else:
                return None
                
        except Exception as e:
            print(f"Error en simulación (a={a:.3f}, b={b:.3f}, c={c:.3f}): {str(e)}")
            return None
    
    def _calculate_hysteretic_energy(self, displacements, moments):
        """Calcula la energía disipada histerética usando el método del trapecio"""
        if len(displacements) < 3 or len(moments) < 3:
            return 0.0
        
        energy = 0.0
        for i in range(1, len(displacements)):
            energy += 0.5 * (moments[i] + moments[i-1]) * (displacements[i] - displacements[i-1])
        
        return abs(energy)
    
    def _calculate_ductility(self, displacements, moments):
        """Calcula la ductilidad como la relación entre desplazamiento último y de fluencia"""
        if len(displacements) < 3:
            return 1.0
            
        # Encontrar punto de fluencia (aproximación: 75% del momento máximo)
        max_moment = max(abs(m) for m in moments)
        yield_moment = 0.75 * max_moment
        
        yield_disp = None
        for i, moment in enumerate(moments):
            if abs(moment) >= yield_moment:
                yield_disp = abs(displacements[i])
                break
        
        if yield_disp is None or yield_disp == 0:
            return 1.0
            
        max_disp = max(abs(d) for d in displacements)
        return max_disp / yield_disp
    
    def generate_dataset(self, target_samples=300, save_progress=True):
        """Genera el dataset completo de combinaciones RBS"""
        print("=== GENERADOR DE DATASET PARA NN2: OPTIMIZACIÓN RBS ===")
        print(f"Objetivo: {target_samples} muestras")
        print(f"Geometría base de viga: L={self.L_beam}m, h={self.h}m, bf={self.bf_total}m")
        print("-" * 60)
        
        # Definir rangos de parámetros
        a_values, b_values, c_values = self.define_parameter_ranges()
        
        print(f"Rangos de parámetros:")
        print(f"  a: {len(a_values)} valores ({a_values.min():.3f} - {a_values.max():.3f} m)")
        print(f"  b: {len(b_values)} valores ({b_values.min():.4f} - {b_values.max():.4f} m)")
        print(f"  c: {len(c_values)} valores ({c_values.min():.3f} - {c_values.max():.3f} m)")
        print(f"  Combinaciones totales posibles: {len(a_values) * len(b_values) * len(c_values)}")
        print("-" * 60)
        
        # Generar combinaciones
        combinations = list(product(a_values, b_values, c_values))
        valid_combinations = []
        
        # Filtrar combinaciones válidas
        for a, b, c in combinations:
            if self.validate_rbs_geometry(a, b, c):
                valid_combinations.append((a, b, c))
        
        print(f"Combinaciones geométricamente válidas: {len(valid_combinations)}")
        
        # Seleccionar muestra representativa si hay demasiadas combinaciones
        if len(valid_combinations) > target_samples:
            # Muestreo estratificado para asegurar diversidad
            indices = np.linspace(0, len(valid_combinations)-1, target_samples, dtype=int)
            selected_combinations = [valid_combinations[i] for i in indices]
        else:
            selected_combinations = valid_combinations
        
        print(f"Combinaciones seleccionadas para simulación: {len(selected_combinations)}")
        print("-" * 60)
        
        # Ejecutar simulaciones
        successful_simulations = 0
        failed_simulations = 0
        start_time = time.time()
        
        for idx, (a, b, c) in enumerate(selected_combinations):
            print(f"Simulando {idx+1}/{len(selected_combinations)} | a={a:.3f}, b={b:.4f}, c={c:.3f}", end=" | ")
            
            # Ejecutar simulación
            result = self.simulate_rbs_performance(a, b, c)
            
            if result is not None:
                # Agregar parámetros de entrada
                result['a'] = a
                result['b'] = b  
                result['c'] = c
                
                # Agregar métricas derivadas
                result['bf_rbs'] = self.bf_total - 2*b
                result['reduction_ratio'] = (2*b) / self.bf_total
                result['rbs_length_ratio'] = (2*c) / self.L_beam
                result['Mp_full'] = self.calculate_plastic_moment(self.bf_total)
                result['Mp_rbs'] = self.calculate_plastic_moment(result['bf_rbs'])
                result['strength_ratio'] = result['Mp_rbs'] / result['Mp_full']
                
                self.dataset.append(result)
                successful_simulations += 1
                print(f"✓ OK (rot={result['max_rotation']:.4f}, M={result['max_moment']/1000:.1f}kNm)")
            else:
                failed_simulations += 1
                print("✗ FALLO")
            
            # Guardar progreso cada 50 simulaciones
            if save_progress and (idx + 1) % 50 == 0:
                self.save_dataset(f"rbs_dataset_progress_{idx+1}.csv")
        
        elapsed_time = time.time() - start_time
        print("-" * 60)
        print(f"RESUMEN DE GENERACIÓN DE DATASET:")
        print(f"  Simulaciones exitosas: {successful_simulations}")
        print(f"  Simulaciones fallidas: {failed_simulations}")
        print(f"  Tasa de éxito: {successful_simulations/(successful_simulations+failed_simulations)*100:.1f}%")
        print(f"  Tiempo total: {elapsed_time/60:.1f} minutos")
        print(f"  Tiempo promedio por simulación: {elapsed_time/len(selected_combinations):.1f} segundos")
        
        return self.dataset
    
    def save_dataset(self, filename="rbs_dataset_complete.csv"):
        """Guarda el dataset en formato CSV"""
        if not self.dataset:
            print("No hay datos para guardar")
            return
        
        df = pd.DataFrame(self.dataset)
        df.to_csv(filename, index=False)
        print(f"Dataset guardado: {filename} ({len(df)} muestras)")
        
        # Mostrar estadísticas del dataset
        print("\nESTADÍSTICAS DEL DATASET:")
        print(df.describe())
    
    def visualize_dataset(self):
        """Crea visualizaciones del dataset generado"""
        if not self.dataset:
            print("No hay datos para visualizar")
            return
        
        df = pd.DataFrame(self.dataset)
        
        fig, axes = plt.subplots(2, 3, figsize=(15, 10))
        fig.suptitle('Análisis del Dataset RBS Generado', fontsize=16)
        
        # 1. Distribución de parámetros a, b, c
        axes[0,0].scatter(df['a'], df['b'], c=df['c'], cmap='viridis', alpha=0.6)
        axes[0,0].set_xlabel('a (m)')
        axes[0,0].set_ylabel('b (m)')
        axes[0,0].set_title('Distribución de Parámetros RBS')
        
        # 2. Rotación vs Momento
        axes[0,1].scatter(df['max_rotation'], df['max_moment']/1000, alpha=0.6)
        axes[0,1].set_xlabel('Rotación Máxima (rad)')
        axes[0,1].set_ylabel('Momento Máximo (kN⋅m)')
        axes[0,1].set_title('Desempeño: Rotación vs Momento')
        
        # 3. Relación de reducción vs desempeño
        axes[0,2].scatter(df['reduction_ratio'], df['max_rotation'], alpha=0.6)
        axes[0,2].set_xlabel('Relación de Reducción')
        axes[0,2].set_ylabel('Rotación Máxima (rad)')
        axes[0,2].set_title('Efecto de la Reducción del Ala')
        
        # 4. Energía disipada
        axes[1,0].hist(df['energy_dissipated'], bins=20, alpha=0.7)
        axes[1,0].set_xlabel('Energía Disipada')
        axes[1,0].set_ylabel('Frecuencia')
        axes[1,0].set_title('Distribución de Energía Disipada')
        
        # 5. Ductilidad
        axes[1,1].hist(df['ductility'], bins=20, alpha=0.7)
        axes[1,1].set_xlabel('Ductilidad')
        axes[1,1].set_ylabel('Frecuencia')
        axes[1,1].set_title('Distribución de Ductilidad')
        
        # 6. Relación de resistencia
        axes[1,2].scatter(df['strength_ratio'], df['max_rotation'], alpha=0.6)
        axes[1,2].set_xlabel('Relación de Resistencia (Mp_rbs/Mp_full)')
        axes[1,2].set_ylabel('Rotación Máxima (rad)')
        axes[1,2].set_title('Resistencia vs Ductilidad')
        
        plt.tight_layout()
        plt.savefig('dataset_analysis.png', dpi=300, bbox_inches='tight')
        plt.show()


# EJECUCIÓN PRINCIPAL
if __name__ == "__main__":
    # Crear generador
    generator = RBSDatasetGenerator()
    
    # Generar dataset
    dataset = generator.generate_dataset(target_samples=300, save_progress=True)
    
    # Guardar dataset final
    generator.save_dataset("rbs_dataset_for_nn2.csv")
    
    # Crear visualizaciones
    generator.visualize_dataset()
    
    print("\n🎯 DATASET LISTO PARA ENTRENAR NN2!")
    print("Archivos generados:")
    print("  - rbs_dataset_for_nn2.csv: Dataset completo")
    print("  - dataset_analysis.png: Visualizaciones del dataset")
```
